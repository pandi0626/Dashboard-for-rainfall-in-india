```python
import dash
from dash import  dcc,html

import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('rainfall in india 1901-2015.csv')
df
```

```python
# Create app layout
app = dash.Dash(__name__)
app.layout = html.Div([
    html.H1('Rainfall in India Dashboard', style={'backgroundColor':'skyblue','textAlign':'center'}),
    
    html.Div([
        html.Label('Select Year:', style={'backgroundColor':'yellow','align':'center'}),
        dcc.Dropdown(
            id='year-dropdown',
            options=[{'label': str(year), 'value': year} for year in df['YEAR'].unique()],
            value=df['YEAR'].min()
        ),
    ]),
    html.Div([
        html.Label('Select state:', style={'backgroundColor':'yellow'}),
        dcc.Dropdown(
            id='state-dropdown',
            options=[{'label': str(subdivision), 'value': subdivision} for subdivision in df['SUBDIVISION'].unique()],
            value=df['SUBDIVISION'].min()
        ),
    ]),
   
     html.Div([
    dcc.Graph(id='bar-chart'),
         
     ], style={'backgroundColor':'black','marginBottom':50,'marginTop':100,'marginleft':300,'padding':12,'width': '70%', 'display': 'inline-block', 'horizontal-align': 'right','vertical-align': 'up',"margin-top":"30px"}),
     html.Div([
    dcc.Graph(id='histogram'),
              ],style={'backgroundColor':'black','marginBottom':60,'marginTop':130,'marginleft':500,'padding':12,'width': '70%', 'display': 'inline-block', 'horizontal-align': 'right','vertical-align': 'up',"margin-top":"30px"}),
      html.Div([
    dcc.Graph(id='scatter-plot'),
      ],style={'backgroundColor':'black','marginBottom':50,'marginTop':100,'marginleft':300,'padding':12,'width': '70%', 'display': 'inline-block', 'horizontal-align': 'right','vertical-align': 'up',"margin-top":"30px"}),
     html.Div([
    dcc.Graph(id='mosaic-plot'),
     ],style={'backgroundColor':'black','marginBottom':50,'marginTop':100,'marginleft':300,'padding':12,'width': '70%', 'display': 'inline-block', 'horizontal-align': 'right','vertical-align': 'up',"margin-top":"30px"}),
             
      html.Div([
        html.Label('Select Year:', style={'backgroundColor':'yellow','align':'center'}),
        dcc.Dropdown(
            id='year-dropdown1',
            options=[{'label': str(year1), 'value': year1} for year1 in df['YEAR'].unique()],
            value=df['YEAR'].min()
        ),
    ]),
    html.Div([ 
    dcc.Graph(id='box-plot'),
    ],style={'backgroundColor':'black','marginBottom':50,'marginTop':100,'marginleft':300,'padding':12,'width': '70%', 'display': 'inline-block', 'horizontal-align': 'right','vertical-align': 'up',"margin-top":"30px"}),
],style={'backgroundColor':'green'}
)
```

```python
# Define callbacks
@app.callback(
    dash.dependencies.Output('bar-chart', 'figure'),
    [dash.dependencies.Input('year-dropdown', 'value')])
def update_bar_chart(value):
    filtered_df = df[df['YEAR'] == value]
    fig = px.bar(filtered_df, x='SUBDIVISION', y='ANNUAL',color='SUBDIVISION')
    fig.update_layout(title='Annual Rainfall by Subdivision')
    return fig

@app.callback(
    dash.dependencies.Output('histogram', 'figure'),
    [dash.dependencies.Input('year-dropdown', 'value')])
def update_histogram(value):
    filtered_df = df[df['YEAR'] == value]
    fig = px.histogram(filtered_df, x='ANNUAL', nbins=30)
    fig.update_layout(title='Distribution of Annual Rainfall')
    return fig

@app.callback(
    dash.dependencies.Output('scatter-plot', 'figure'),
    [dash.dependencies.Input('year-dropdown', 'value')])
def update_scatter_plot(value):
    filtered_df = df[df['YEAR'] == value]
    fig = px.scatter(filtered_df, x='JAN', y='FEB', size='ANNUAL', color='SUBDIVISION')
    fig.update_layout(title='Monthly Rainfall Scatter Plot')
    return fig

@app.callback(
    dash.dependencies.Output('mosaic-plot', 'figure'),
    [dash.dependencies.Input('year-dropdown', 'value')])
def update_mosaic_plot(value):
    filtered_df = df[df['YEAR'] == value]
    fig = px.imshow(filtered_df.pivot_table(index='YEAR', columns='SUBDIVISION', values='ANNUAL'),
                     color_continuous_scale='oxy', title='Rainfall Mosaic Plot')
    return fig

@app.callback(
    dash.dependencies.Output('box-plot', 'figure'),
    [dash.dependencies.Input('year-dropdown1', 'value')])
def update_box_plot(value):
    filtered_df = df[df['YEAR'] == value]
    fig = px.box(filtered_df,  y='Mar-May', color='YEAR')
    fig.update_layout(title='Mar-May Rainfall by Subdivision')
    return fig
```

```python

if __name__ == '__main__':
    
    app.run_server(port=8080)
```
